The averages are computed as: 
$$0.3Q + 0.3L + 0.4T = Ave$$

```python
grades['Ave'] = (grades*weights).sum(axis=1)
grades
```

### Binning

```python
bins = [0,60,70,80,90,100]
letter_grades = ['F','D','C','B','A']
```

```python
gradebook
```

```python
gradebook['grade'] = pd.cut(gradebook.Ave,bins,labels=letter_grades)
gradebook
```

### Grouping

```python
gradebook
```

```python
df = gradebook
pd.crosstab(df.major,df.year)
```

```python
df = gradebook
pd.crosstab(df.major,df.year,margins=True)
```

```python
df = gradebook
pd.crosstab([df.major,df.year],df.gender)
```

```python
gradebook
```

```python
df = gradebook
groupings = df.groupby([df.major,df.year])
groupings.mean()
```

```python
df = gradebook
groupings = df.groupby(df.major)
groupings['Ave'].agg(['count','max','min'])
```
