```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.svm import SVR
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_validate, GridSearchCV
%matplotlib inline
```

```python
df = pd.read_csv("./Concrete_train.csv")
df.head()
```

```python
features_train = df.drop(df.columns[-1],axis=1)
target_train = df.iloc[:,-1]
```

```python
features_train = (features_train - features_train.mean()) / features_train.std()
features_train.head()
```

```python
grid = {'C': np.linspace(100,400,10)}

clf = SVR()
clfCV = GridSearchCV(clf,param_grid=grid,return_train_score=True,n_jobs=-1)
clfCV.fit(features_train,target_train)

print()
print('parameters = ', clfCV.best_params_, 'validation R2 = ', clfCV.best_score_.round(3))
```
