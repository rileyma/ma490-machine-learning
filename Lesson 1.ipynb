## (F) What is the biggest margin of victory in a superbowl?

```python
df['Win Margin'] = df.Pts - df['Pts.1']
df['Win Margin'].max()
```

ANS: 45 points

## (G) What is the lowest total number of points scored in a superbowl?

```python
df['Total Points'] = df.Pts + df['Pts.1']
df['Total Points'].min()
```

ANS: 16 points
